# Feature Engineering

## Creamos la columna de mood según energy y valence

```python
#  funcion para clasificar canciones por felicidad segun su energy
def clasificar_mood(row):
    if row["valence"] > 0.6 and row["energy"] > 0.6:
        return "happy"
    elif row["valence"] < 0.4 and row["energy"] < 0.5:
        return "sad"
    elif row["energy"] > 0.7 and row["valence"] < 0.5:
        return "angry"
    else:
        return "neutral"

df["mood_inferido"] = df.apply(clasificar_mood, axis=1)

df["mood_inferido"].value_counts()
```

```python
df.to_csv("../data/processed/songs_clean.csv", index=False)
print(" Archivo final guardado en data/processed/songs_clean.csv")
```
